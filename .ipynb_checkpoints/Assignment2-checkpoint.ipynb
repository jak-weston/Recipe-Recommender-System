```python
import os
import pandas as pd
import kagglehub
import matplotlib.pyplot as plt
import seaborn as sns
```

### Download the dataset

```python
path = kagglehub.dataset_download("shuyangli94/food-com-recipes-and-user-interactions")
print("Path to dataset files:", path)

# List all files in the dataset directory
dataset_files = os.listdir(path)
print("Files in the dataset:", dataset_files)
```

### Create dictionary for each dataset

```python
# Dictionary to store DataFrames for each useful file
dataframes = {}

# Load each useful CSV file into a DataFrame dictionary
for file_name in dataset_files:
    file_path = os.path.join(path, file_name)
    if file_name.endswith('.csv') and os.path.exists(file_path):
        print(f"Loading {file_name} into a DataFrame...")
        df_name = file_name.split('.')[0]  # Use filename without extension as key
        dataframes[df_name] = pd.read_csv(file_path)
        print(f"Loaded {file_name} with {dataframes[df_name].shape[0]} rows and {dataframes[df_name].shape[1]} columns.")
```

### Example: Access the RAW_recipes DataFrame

```python
raw_recipes_df = dataframes.get('RAW_recipes')
raw_recipes_df.head()
```

# EDA: Recipes Dataset

### Top 10 Most Common Tags

```python
tag_counts = raw_recipes_df['tags'].str.strip("[]").str.replace("'", "").str.split(", ").explode().value_counts()
top_tags = tag_counts.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_tags.values, y=top_tags.index)
plt.title("Top 10 Most Common Recipe Tags", fontsize=16)
plt.xlabel("Frequency", fontsize=12)
plt.ylabel("Tags", fontsize=12)
plt.tight_layout()
plt.show()
```

### Number of Ingredients vs. Number of Steps

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='n_ingredients', y='n_steps', data=raw_recipes_df, alpha=0.7)
plt.title("Number of Ingredients vs. Number of Steps", fontsize=16)
plt.xlabel("Number of Ingredients", fontsize=12)
plt.ylabel("Number of Steps", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Rename the 'id' column in raw_recipes_df to match the 'recipe_id' column in RAW_interactions_df
raw_recipes_df = raw_recipes_df.rename(columns={'id': 'recipe_id'})

# Merge the datasets
merged_df = pd.merge(raw_recipes_df, dataframes['RAW_interactions'], on='recipe_id', how='inner')

# Verify the merge
print("Merged DataFrame shape:", merged_df.shape)
merged_df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(merged_df['rating'], bins=10, kde=True, color='blue', alpha=0.7)
plt.title("Distribution of Recipe Ratings", fontsize=16)
plt.xlabel("Rating", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='minutes', y='rating', data=merged_df, alpha=0.7)
plt.title("Cooking Time vs. Rating", fontsize=16)
plt.xlabel("Cooking Time (minutes)", fontsize=12)
plt.ylabel("Rating", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='rating', y='n_ingredients', data=merged_df)
plt.title("Number of Ingredients vs. Rating", fontsize=16)
plt.xlabel("Rating", fontsize=12)
plt.ylabel("Number of Ingredients", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import ast

# Convert string representations of lists into actual lists
raw_recipes_df['nutrition'] = raw_recipes_df['nutrition'].apply(ast.literal_eval)

# Verify the conversion
print("Converted nutrition values:", raw_recipes_df['nutrition'].head())
print("Data type of nutrition column after conversion:", type(raw_recipes_df['nutrition'][0]))

# Define column names for the nutrition data
nutrition_columns = ['calories', 'total_fat', 'sugar', 'sodium', 'protein', 'saturated_fat']

# Expand the 'nutrition' lists into separate columns
nutrition_data = pd.DataFrame(raw_recipes_df['nutrition'].tolist(), columns=nutrition_columns)

# Add the new columns back to the original DataFrame
raw_recipes_df = pd.concat([raw_recipes_df, nutrition_data], axis=1)

# Verify the result
print("Columns in raw_recipes_df after expansion:", raw_recipes_df.columns)
print(raw_recipes_df[['calories', 'total_fat', 'sugar', 'sodium', 'protein', 'saturated_fat']].head())
plt.figure(figsize=(10, 6))
sns.histplot(raw_recipes_df['calories'], bins=30, kde=True, color='blue', alpha=0.7)
plt.title("Distribution of Calories in Recipes", fontsize=16)
plt.xlabel("Calories", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud

# Flatten ingredient lists and join as a single string
ingredient_text = ' '.join(merged_df['ingredients'].dropna().sum())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(ingredient_text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Most Common Ingredients", fontsize=16)
plt.show()
```
