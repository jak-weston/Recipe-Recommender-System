```python
import os
import pandas as pd
import kagglehub
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from collections import Counter  # Ensure this import is included
import ast
import matplotlib.cm as cm
import cupy as cp
import ast  # To safely evaluate string representations of lists
```

### Download the dataset

```python
path = kagglehub.dataset_download("shuyangli94/food-com-recipes-and-user-interactions")
print("Path to dataset files:", path)

# List all files in the dataset directory
dataset_files = os.listdir(path)
print("Files in the dataset:", dataset_files)
```

### Create dictionary for each dataset

```python
# Dictionary to store DataFrames for each useful file
dataframes = {}

# Load each useful CSV file into a DataFrame dictionary
for file_name in dataset_files:
    file_path = os.path.join(path, file_name)
    if file_name.endswith('.csv') and os.path.exists(file_path):
        print(f"Loading {file_name} into a DataFrame...")
        df_name = file_name.split('.')[0]  # Use filename without extension as key
        dataframes[df_name] = pd.read_csv(file_path)
        print(f"Loaded {file_name} with {dataframes[df_name].shape[0]} rows and {dataframes[df_name].shape[1]} columns.")
```

### Example: Access the RAW_recipes DataFrame

```python
raw_recipes_df = dataframes.get('RAW_recipes')
raw_recipes_df.head()
```

```python
# Load your pp_recipes dataframe
pp_recipes_df = dataframes.get('PP_recipes')
pp_recipes_df.head()
```

```python
if 'techniques' in pp_recipes_df.columns:
    techniques_types = pp_recipes_df['techniques'].apply(type).value_counts()
    print("Types of data stored in the 'techniques' column:")
    print(techniques_types)
else:
    print("The 'techniques' column does not exist in the DataFrame.")
```

```python
# Convert techniques strings to lists of integers
try:
    pp_recipes_df['techniques'] = pp_recipes_df['techniques'].apply(ast.literal_eval)
    print("Successfully converted 'techniques' strings to lists.")
except Exception as e:
    print(f"Error during conversion: {e}")
```

```python
# Ensure 'techniques' column exists and is in the correct format
if 'techniques' not in pp_recipes_df.columns:
    raise ValueError("The dataframe must contain a 'techniques' column.")

# Define the list of techniques
TECHNIQUES_LIST = [
    'bake', 'barbecue', 'blanch', 'blend', 'boil', 'braise', 'brine', 'broil', 'caramelize', 'combine',
    'crock pot', 'crush', 'deglaze', 'devein', 'dice', 'distill', 'drain', 'emulsify', 'ferment', 'freez',
    'fry', 'grate', 'griddle', 'grill', 'knead', 'leaven', 'marinate', 'mash', 'melt', 'microwave',
    'parboil', 'pickle', 'poach', 'pour', 'pressure cook', 'puree', 'refrigerat', 'roast', 'saute',
    'scald', 'scramble', 'shred', 'simmer', 'skillet', 'slow cook', 'smoke', 'smooth', 'soak', 'sous-vide',
    'steam', 'stew', 'strain', 'tenderize', 'thicken', 'toast', 'toss', 'whip', 'whisk'
]

# Initialize a Counter for techniques
technique_counts = Counter()

# Loop through each row in the 'techniques' column
for technique_row in pp_recipes_df['techniques']:
    if isinstance(technique_row, (list, tuple)):
        for index, value in enumerate(technique_row):
            if value == 1:  # Check for the presence of a technique
                technique_counts[TECHNIQUES_LIST[index]] += 1

# Extract data for the pie chart
labels = list(technique_counts.keys())
sizes = list(technique_counts.values())

# Use a colormap for better colors
cmap = cm.get_cmap('tab20c')
colors = cmap.colors[:len(labels)]

# Calculate percentages
total_techniques = sum(sizes)
percentages = [f"{(size / total_techniques) * 100:.1f}%" for size in sizes]

# Plot the pie chart
plt.figure(figsize=(12, 8))
wedges, texts, autotexts = plt.pie(
    sizes, 
    labels=None,  # Exclude labels from the pie chart
    autopct='%1.1f%%', 
    startangle=140, 
    colors=colors
)

# Add a legend with the technique names and percentages
legend_labels = [f"{label}: {percentage}" for label, percentage in zip(labels, percentages)]
plt.legend(
    wedges, 
    legend_labels, 
    title="Techniques", 
    loc="center left", 
    bbox_to_anchor=(1, 0.5),
    fontsize=10
)

# Title and formatting
plt.title('Distribution of Techniques in Recipes', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular
plt.show()
```
